**Computing the value of RUL (Remaining Unit Life)**

RUL = Failure cycle - Current cycle

1. Creating a column called failure_cycle

```python
dataset["failure_cycle"] = dataset.groupby("Unit_Number")["time(in_cycles)"].transform("max")
dataset[["Unit_Number" , "time(in_cycles)" , "failure_cycle"]]
```

2. Creating the RUL Column (The Target Variable)

```python
dataset["RUL"] = dataset["failure_cycle"] - dataset["time(in_cycles)"]
dataset[["Unit_Number" , "time(in_cycles)" , "failure_cycle" , "RUL"]]
```

```python
dataset
```

3. Dropping the failure_cycle after computing RUL (as we want the model to learn the sensor degradation pattens)

```python
dataset = dataset.drop(columns = ["failure_cycle"])
```

Creating the Sequential data for the model training


```python
def create_sequences_per_engine(data, seq_length=30):

    X_seq = []
    y_seq = []

    for engine_id in data["Unit_Number"].unique():

        engine_data = data[data["Unit_Number"] == engine_id]

        X = engine_data.drop(["Unit_Number","RUL"], axis=1).values
        y = engine_data["RUL"].values

        for i in range(len(X) - seq_length):
            X_seq.append(X[i:i+seq_length])
            y_seq.append(y[i+seq_length])

    return np.array(X_seq), np.array(y_seq)
```

```python
X_seq , y_seq = create_sequences_per_engine(dataset_dropped , seq_length = 30)
```

```python
print(X_seq)
```

```python
print(y_seq)
```

```python
print(X_seq.shape)
print(y_seq.shape)                        # We lose 30 rows per engine(total 100) to predict the RUL based on the values from the last 30 cycles
```

```python
from sklearn.model_selection import train_test_split

X_train_seq , X_val_seq , y_train_seq , y_val_seq = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)
```

Creating the model

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN , Dense , LSTM , Dropout
```

***RNN***

```python
rnn_model = Sequential([
    SimpleRNN(units = 64 , activation = "tanh" , return_sequences=True , input_shape = (30 , X_train_seq.shape[2])),
    SimpleRNN(units = 32),
    Dense(units = 32 , activation = "relu"),
    Dense(units = 1)                                                   # Default activation function - linear
])

rnn_model.compile(optimizer = "adam" , loss = "mse")
```

```python
rnn_model_history = rnn_model.fit(X_train_seq , y_train_seq , 
                                  validation_data = (X_val_seq , y_val_seq) , 
                                  epochs = 20 , batch_size = 64)
```

```python
y_pred_rnn = rnn_model.predict(X_val_seq)
```

```python
rmse_rnn = root_mean_squared_error(y_val_seq, y_pred_rnn)
r2_rnn = r2_score(y_val_seq, y_pred_rnn)

print("RNN RMSE:", rmse_rnn)
print("RNN R2:", r2_rnn)
```

**LSTM**

```python
lstm_model = Sequential([
    LSTM(units = 48 , input_shape = (30 , X_train_seq.shape[2])),
    Dropout(0.2),
    Dense(units = 32 , activation = "relu"),
    Dense(units = 1)                                                   # Default activation function - linear
])

lstm_model.compile(optimizer = "adam" , loss = "mse")
```

```python
lstm_model_history = lstm_model.fit(X_train_seq , y_train_seq , 
                                  validation_data = (X_val_seq , y_val_seq) , 
                                  epochs = 20 , batch_size = 64)
```

```python
y_pred_lstm = lstm_model.predict(X_val_seq)
```

```python
rmse_lstm = root_mean_squared_error(y_val_seq, y_pred_lstm)
r2_lstm = r2_score(y_val_seq, y_pred_lstm)

print("LSTM RMSE:", rmse_lstm)
print("LSTM R2:", r2_lstm)
```

```python
def create_test_sequences(data, seq_length=30):

    X_test_seq = []

    for engine_id in data["Unit_Number"].unique():

        engine_data = data[data["Unit_Number"] == engine_id]

        X = engine_data.drop(["Unit_Number"], axis=1).values

        if len(X) >= seq_length:
            X_test_seq.append(X[-seq_length:])   # last 30 cycles

        else:
            padding = np.zeros((seq_length - len(X), X.shape[1]))
            X_test_seq.append(np.vstack((padding, X)))

    return np.array(X_test_seq)
```
